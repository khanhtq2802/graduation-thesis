```python
# load loss history from file
training_losses = []
test_losses = []
try:
    with open("losses.txt", "r") as f:
        for line in f:
            if line.strip():  # Check if line is not empty
                training_loss, test_loss = line.split(",")
                training_losses.append(float(training_loss))
                test_losses.append(float(test_loss))
except:
    print("error when load training history")

plt.plot(training_losses[-50:], label="Training Loss")
plt.plot(test_losses[-50:], label="test Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
# save loss history to file
with open("losses.txt", "w") as f:
    for i in range(len(training_losses)):
        f.write(f"{training_losses[i]},{test_losses[i]}\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D liu
polarities = ['None/Neutral', 'Positive', 'Negative', 'Conflict']
f1_scores = {
    'all': [0.678, 0.9001, 0.8989, 0.6907],
    'amn': [0.9374, 0.7911, 0.7087, 0.32],
    'ch': [0.9331, 0.7928, 0.8571, 0.6087],
    'mgt': [0.9539, 0.6813, 0.7738, 0.3529],
    'nat': [0.964, 0.8187, 0.8, 0.5455],
    'ppl': [0.9746, 0.8, 0.6222, 0.1667]
}

# Thit lp biu 
fig, ax = plt.subplots(figsize=(11, 8))

# S lng polarities
num_polarities = len(polarities)
index = np.arange(num_polarities)

# Chiu rng ca mi nhm bars
bar_width = 0.15

# V tng nhm bars cho mi aspect category
for i, (aspect_category, scores) in enumerate(f1_scores.items()):
    bars = ax.bar(index + i * bar_width, scores, bar_width, label=aspect_category)

    # Adding annotations on top of each bar
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=7)  # Set font size to 8

# nh dng biu 
ax.set_xlabel('Polarity')
ax.set_ylabel('F1 Score')
ax.set_title('F1 Scores by Polarity and Aspect Category')
ax.set_xticks(index + bar_width * (len(f1_scores) - 1) / 2)
ax.set_xticklabels(polarities)
ax.legend()

# Hin th biu 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
aspect_categories = ['All', 'Amn', 'Ch', 'Mgt', 'Nat', 'Ppl']
macro_f1_scores = [0.8256, 0.8124, 0.7178, 0.8030, 0.8609, 0.7990]

# Sorting the data by macro_f1_scores in ascending order
sorted_indices = np.argsort(macro_f1_scores)
sorted_aspect_categories = [aspect_categories[i] for i in sorted_indices]
sorted_macro_f1_scores = [macro_f1_scores[i] for i in sorted_indices]

# Plotting
plt.figure(figsize=(7, 6))
bars = plt.bar(sorted_aspect_categories, sorted_macro_f1_scores, color=['#2ca02c', '#8c564b', '#d62728', '#ff7f0e', '#1f77b4', '#9467bd'])

# Adding title and labels
plt.title('Comparison of Aspect Categories based on Macro-F1 Scores')
plt.xlabel('Aspect Categories')
plt.ylabel('Macro-F1 Scores')

# Adding value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.4f}', ha='center', va='bottom')

# Display the plot
plt.ylim(0, 1)
plt.show()
```
